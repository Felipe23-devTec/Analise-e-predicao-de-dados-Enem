```python
import pandas as pd

tabelaBelo = pd.read_csv('dadosBeloJardim.csv',delimiter=',',encoding='ISO-8859-1')
display(tabelaBelo)

tabelaBelo[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT','NU_NOTA_REDACAO']] = tabelaBelo[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT','NU_NOTA_REDACAO']].fillna(0)

display(tabelaBelo)
```

```python
#Transformando dados
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
tabelaBelo['Q025'] = tabelaBelo['Q025'].replace({'A': 0, 'B': 1})
display(tabelaBelo['Q025'])
tabelaBelo['Q024'] = tabelaBelo['Q024'].replace({'A': 0, 'B': 1,'C': 1, 'D': 1, 'E': 1})
display(tabelaBelo['Q024'])
cont = tabelaBelo['Q024'].value_counts()
print(cont)

tabelaBelo['NU_NOTA_TOTAL'] = tabelaBelo[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT','NU_NOTA_REDACAO']].mean(axis=1)
tabelaBelo['DESEMPENHO'] = tabelaBelo.apply(lambda x: 1 if(x['NU_NOTA_TOTAL'] >= 500) and
                                                                       (x['Q025'] == 1) and 
                                                                       (x['Q024'] == 1) 
                                                                       else 0, axis=1)

print("Tabela de desenpenho: ")
print(tabelaBelo['DESEMPENHO'].value_counts())

X = tabelaBelo[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT','NU_NOTA_REDACAO','Q025','Q024']]
y = tabelaBelo['DESEMPENHO']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = GaussianNB()
model.fit(X_train, y_train)

# 5. Testar o modelo
y_pred = model.predict(X_test)

# 6. Avaliar o desempenho do modelo
acc = accuracy_score(y_test, y_pred)
print(f'Acurácia: {acc}')


novo_estudante = pd.DataFrame({
    'NU_NOTA_CN': [400],
    'NU_NOTA_CH': [400],
    'NU_NOTA_LC': [400],
    'NU_NOTA_MT': [400],
    'NU_NOTA_REDACAO': [400],
    'Q025': [1],
    'Q024': [1]
})

# Prever o desempenho do estudante com o modelo treinado
probabilidade = model.predict_proba(novo_estudante)[:,1]

# Exibir a probabilidade de desempenho satisfatório do estudante
print(f"A probabilidade de desempenho satisfatório do estudante é de {probabilidade[0]*100:.2f}%.")

```
