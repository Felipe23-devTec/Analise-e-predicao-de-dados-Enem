```python
#Importação de bibliotecas
import pandas as pd
```

```python
#parte relacionado ao filtro de todos os dados do enem com filtro para belo jardim
tabela = pd.read_csv('MICRODADOS_ENEM_2021.csv',delimiter=';',encoding='ISO-8859-1')
filtro = (tabela['NO_MUNICIPIO_ESC'] == 'Belo Jardim') & (tabela['NO_MUNICIPIO_PROVA'] == 'Belo Jardim')

nova_tabela = tabela[filtro]
nova_tabela = nova_tabela.dropna()
display(nova_tabela)
nova_tabela.to_csv("dadosBeloJardim2021.csv", index=False)

#Mesclagem de base de dados relacionados aos anos do enem

arquivo1 = 'dadosBeloJardim2019.csv'
arquivo2 = 'dadosBeloJardim2020.csv'
arquivo3 = 'dadosBeloJardim2021.csv'
arquivo4 = 'dadosBeloJardim2022.csv'
df1 = pd.read_csv(arquivo1)
df2 = pd.read_csv(arquivo2)
df3 = pd.read_csv(arquivo3)
df4 = pd.read_csv(arquivo4)
df_concatenado = pd.concat([df1, df2,df3,df4])

arquivo_concatenado = 'dadosBeloJardim.csv'
df_concatenado.to_csv(arquivo_concatenado, index=False)
```
